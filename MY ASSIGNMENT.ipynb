```python
#calculate_average(marks)

def calculate_average(marks):
    return sum(marks) / len(marks)
```

```python
#determine_grade(average) function
def determine_grade(average):
    if average >= 70:
        return "A"
    elif average >= 60:
        return "B"
    elif average >= 50:
        return "C"
    elif average >= 40:
        return "D"
    else:
        return" Fail"
```
