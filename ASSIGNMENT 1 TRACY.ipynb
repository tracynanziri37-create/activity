```python
class Book:
    def __init__(self, title, author, year):
        self.title = title
        self.author = author
        self.year = year

    def display_details(self):
        print(f"Title: {self.title}")
        print(f"Author: {self.author}")
        print(f"Year: {self.year}")

my_book = Book ("Python Ethics", "Tracy", 2005)
my_book.display_details()
        
    
```

```python
class car:
    def __init__(self, brand, model, year):
        self.brand = brand
        self.model = model
        self.year = year

my_car = car("Benz", "Gwagen", 2020)

print(my_car.brand)
print(my_car.model)
print(my_car.year)
```

```python
class Light:
    def __init__(self, color="white", is_on=False):
        self.color = color
        self.is_on = bool(is_on)

lamp = Light("warm", True)
lamp.is_on = not lamp.is_on
print(lamp.is_on) 
```

```python
class Order:
    def __init__(self, order_id, items):
        self.order_id = order_id; self.items = list(items)
    def __str__(self):
        return f"Order {self.order_id}: {len(self.items)} items"

print(Order("ORD100", ["pen", "notebook"]))  
```
